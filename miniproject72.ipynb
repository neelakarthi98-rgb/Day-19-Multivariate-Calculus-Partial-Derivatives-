## Visualizing a Gradient on a Contour Plot

////Objective: Create a 2D contour plot of a function like f(x,y)=x2+y2 and use a quiver plot to visualize the gradient vector at a few points

```python
import numpy as np
import matplotlib.pyplot as plt
```

///Define a function and its gradient//

```python
def f(x,y):
    return x**2+y**2
def grad_f(x,y):
    df_dx=2*x
    df_dy=2*y
    return df_dx,df_dy
```

///craete a grid of(x,y)points

```python
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

////compute gradient vectors///

```python
U, V = grad_f(X, Y)  # U = ∂f/∂x, V = ∂f/∂y
```

////Create contour plot///

```python
plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')
plt.clabel(contours, inline=True, fontsize=8)
plt.title("Contour and Gradient (Quiver) Plot of f(x, y) = x² + y²")
```

/// Overlay quiver (gradient vectors)

```python
plt.quiver(X, Y, U, V, color='black', scale=50)


plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
