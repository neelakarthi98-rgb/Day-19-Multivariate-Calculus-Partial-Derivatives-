## Calculating Partial Derivatives Manually

////Objective: Write a Python function to compute the partial derivative of a simple multivariate function.///

```python
def f(x,y):
    return 2*x**2+3*x*y+y**3
def partial_x(x,y):
    return 4*x+3*y
def partial_y(x,y):
    return 3*x+3*y**2
```

///Assign a value x=1,y=2 to calculate partial  derivative//

```python
x_val=1
y_val=2
f_val=f(x_val,y_val)
df_dx=partial_x(x_val,y_val)
df_dy=partial_y(x_val,y_val)
```

////print results//

```python
print("function:f(x,y)=2x^2+3xy+y^3")
print(f"At point(x,y)=({x_val},{y_val})")
print(f"F(X,Y)={f_val}")
print(f"df_dx ={df_dx}")
print(f"df_dy ={df_dy}")
```
